```python
import matplotlib.pyplot as plt
import numpy as np
from Environment import environment
```

```python
data = []
pre_normalisation_cdata = []
pre_normalisation_bdata = []
c = []
b = []

cvalues=np.linspace(2, 4, 100)
bvalues=np.linspace(40, 65, len(cvalues))
cdata=np.linspace(2, 4, len(cvalues))
bdata=np.linspace(40, 65, len(cvalues))

for i in range(len(bdata)):

    cvalues[i] = (cvalues[i] - 2) / 2
    bvalues[i] = (bvalues[i] - 45) / 20
    c.append(cvalues[i])
    b.append(bvalues[i])
    func=environment(b[i], c[i])
    data.append(float(func))
```

```python
data=list(data)
plt.ylabel('Lap Shear Strength (N)')
plt.xlabel('Clamping pressure (bar) and vibration amplitude(um)')
plt.plot(cdata,data)
plt.show()
```
